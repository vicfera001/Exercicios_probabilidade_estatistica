Exercício 9

```python
import matplotlib.pyplot as plt

# Dados para o gráfico de pizza
labels = ['Maçãs', 'Bananas', 'Laranjas', 'Morangos']
quantidades = [30, 45, 25, 20]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Criação do gráfico de pizza
plt.pie(quantidades, labels=labels, colors=colors, autopct='%1.1f%%',startangle=90)

# Adiciona um círculo no meio para transformar em um gráfico de pizza
centro_circuito = plt.Circle((0, 0), 0.7, color='white')
fig = plt.gcf()
fig.gca().add_artist(centro_circuito)

# Ajustes estéticos
plt.axis('equal') # _Para garantir que o gráfico seja exibido como um círculo_
plt.title('Distribuição de Frutas')

# Mostra o gráfico de pizza
plt.show()
```
